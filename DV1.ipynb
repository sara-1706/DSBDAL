```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(12,6))
sns.histplot( data= df, x='fare', bins=30,hue = 'class', kde= True, palette='bright')
plt.title("Mera mann", fontsize = 14)
plt.xlabel("Fare", fontsize= 14)
plt.ylabel("No of passengers", fontsize = 14)
plt.grid(True)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="sex", y="age",hue="survived",data= df, palette='coolwarm')
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(x="survived", y="age", hue="sex", palette="rainbow", data= df)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="survived", y= "age", hue ="sex", data=df, palette="rainbow")
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x="survived", y="age", hue="sex" , data= df, palette = "rainbow")
```
